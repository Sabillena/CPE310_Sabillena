Exercise #1

```python
import pandas as pd

meteorites = pd.read_csv("/content/Meteorite_Landings.csv")

meteorites.rename(columns={'mass (g)': 'mass'}, inplace=True)
meteorites.drop(columns=['latitude', 'longitude'], errors='ignore', inplace=True)
meteorites_sorted = meteorites.sort_values(by='mass', ascending=False)
meteorites_sorted.head()
```

Exercise #2

```python
meteorites['year'] = meteorites['year'].astype(str).str.slice(0, 4)
meteorites['year'] = pd.to_numeric(meteorites['year'], errors='coerce')
meteorites['pre_1970'] = meteorites['year'] < 1970

meteorites.set_index('id', inplace=True)
meteorites.sort_index(inplace=True)

selected_rows = meteorites.loc[10036:10040]
selected_rows
```

Exercise #3
A.

```python
meteorites['year'] = pd.to_numeric(meteorites['year'], errors='coerce')

filtered = meteorites[(meteorites['year'] >= 2005) & (meteorites['year'] <= 2009)]

pivot_table = filtered.pivot_table(
    index='year',
    columns='fall',
    values='mass',
    aggfunc=[len, lambda x: x.quantile(0.95)]
)

pivot_table
```

Exercise #3 B.

```python
fall_stats = meteorites.groupby('fall')['mass'].describe()
fall_stats
```

```python
taxi = pd.read_csv("/content/2019_Yellow_Taxi_Trip_Data.csv", parse_dates=['tpep_dropoff_datetime'])

taxi.set_index('tpep_dropoff_datetime', inplace=True)
hourly = taxi.resample('H').agg({
    'trip_distance': 'sum',
    'fare_amount': 'sum',
    'tolls_amount': 'sum',
    'tip_amount': 'sum'
})

top_tips = hourly.sort_values(by='tip_amount', ascending=False).head(5)
top_tips
```
